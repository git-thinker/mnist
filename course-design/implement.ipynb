```python
index = {int(i.split()[1][1:]) :  i for i in os.listdir('vitNet') if '.pt' in i}
aca = []

for i in range(1, len(index)+1):
    max_ = 0
    max__ = ''
    net = VitNet().cuda()
    net.load_state_dict(torch.load(os.path.join('vitNet', index[i]), map_location='cuda'))
    net.eval()
    alls = 0
    corrects = 0
    for im, labels, path in devloader:
        im, labels = im.to('cuda'), labels.to('cuda')
        y = net(im.view([-1, *im.shape[2:]])).argmax(dim=1).view([-1, 6])
        alls += y.shape[-1]
        corrects += (y==labels).int().sum()
        # for i in range(6):
        #     n = plt.subplot(1, 6, i+1)
        #     n = plt.imshow(im[0][i].squeeze(),cmap='gray')
        # plt.show()
    print("---------------------")
    print(i)
    print(corrects / alls)
    aca.append(corrects / alls)
    if (corrects / alls < max_):
        max_ = corrects / alls
        max__ = path
```
